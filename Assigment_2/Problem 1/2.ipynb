```python
import numpy as np
from scipy.stats import gamma

# Given data
X = np.array([5, 6, 4])

# Part (a) - Calculate the MLE
N = len(X)
x_bar = np.mean(X)
theta_mle = 1 / x_bar
print(f'MLE of theta: {theta_mle}')

# Part (b) - MLE given data
print(f'MLE given data: {theta_mle}')

# Part (c) - Choose the prior parameter λ̂ such that E[θ] = 1/3
lambda_hat = 3
print(f'λ̂: {lambda_hat}')

# Part (d) - Posterior distribution parameters
a_post = N + 1
b_post = np.sum(X) + lambda_hat

# Part (f) - Posterior mean
posterior_mean = a_post / b_post
print(f'Posterior mean of theta: {posterior_mean}')

# Using scipy.stats.gamma to verify the posterior distribution
posterior_distribution = gamma(a=a_post, scale=1/b_post)
print(f'Posterior mean using scipy: {posterior_distribution.mean()}')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma, expon

# Given data
X = np.array([5, 6, 4])

# Part (a) - Calculate the MLE
N = len(X)
x_bar = np.mean(X)
theta_mle = 1 / x_bar

# Part (b) - MLE given data
print(f'MLE given data: {theta_mle}')

# Part (c) - Choose the prior parameter λ̂ such that E[θ] = 1/3
lambda_hat = 3
print(f'λ̂: {lambda_hat}')

# Part (d) - Posterior distribution parameters
a_post = N + 1
b_post = np.sum(X) + lambda_hat

# Part (f) - Posterior mean
posterior_mean = a_post / b_post
print(f'Posterior mean of theta: {posterior_mean}')

# Using scipy.stats.gamma to verify the posterior distribution
posterior_distribution = gamma(a=a_post, scale=1/b_post)
print(f'Posterior mean using scipy: {posterior_distribution.mean()}')

# Plotting
theta_values = np.linspace(0.01, 1, 500)
prior = expon(scale=1/lambda_hat).pdf(theta_values)
likelihood = theta_values**N * np.exp(-theta_values * np.sum(X))
posterior = posterior_distribution.pdf(theta_values)

# Normalize likelihood for plotting
likelihood /= np.max(likelihood)

# Plotting the prior distribution
plt.figure(figsize=(10, 6))
plt.plot(theta_values, prior, label='Prior: Exponential(λ=3)', color='blue')
plt.xlabel('θ')
plt.ylabel('Density')
plt.title('Prior Distribution')
plt.legend()
plt.grid(True)
plt.show()

# Plotting the likelihood function
plt.figure(figsize=(10, 6))
plt.plot(theta_values, likelihood, label='Likelihood', color='green')
plt.xlabel('θ')
plt.ylabel('Density')
plt.title('Likelihood Function')
plt.legend()
plt.grid(True)
plt.show()

# Plotting the posterior distribution
plt.figure(figsize=(10, 6))
plt.plot(theta_values, posterior, label='Posterior: Gamma(a=4, b=18)', color='red')
plt.xlabel('θ')
plt.ylabel('Density')
plt.title('Posterior Distribution')
plt.legend()
plt.grid(True)
plt.show()
```
