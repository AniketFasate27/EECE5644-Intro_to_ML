```python
import numpy as np
import matplotlib.pyplot as plt

# Initialize data
n = 3  # number of initial data points
x = np.array([1, 2, 3])  # example x values
y = np.array([2, 3, 4])  # example y values

# Calculate initial statistics
x_bar = np.mean(x)
y_bar = np.mean(y)
C_xx = np.mean((x - x_bar) ** 2)
C_xy = np.mean((x - x_bar) * (y - y_bar))

# New data point
x_new = 4
y_new = 5

# Update statistics
x_bar_new = x_bar + (x_new - x_bar) / (n + 1)
y_bar_new = y_bar + (y_new - y_bar) / (n + 1)
C_xx_new = (1 / (n + 1)) * (x_new**2 + n * C_xx + n * x_bar**2 - (n + 1) * x_bar_new**2)
C_xy_new = (1 / (n + 1)) * (x_new * y_new + n * C_xy + n * x_bar * y_bar - (n + 1) * x_bar_new * y_bar_new)

# Print updated statistics
print(f'Updated x_bar: {x_bar_new}')
print(f'Updated y_bar: {y_bar_new}')
print(f'Updated C_xx: {C_xx_new}')
print(f'Updated C_xy: {C_xy_new}')

# Plotting the data
# Plot initial x and y values
plt.figure(figsize=(10, 6))
plt.scatter(x, y, label='Initial data points', color='blue')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Initial Data Points')
plt.legend()
plt.grid(True)
plt.savefig('initial_data.png')
plt.show()

# Plot updated x and y values
x_updated = np.append(x, x_new)
y_updated = np.append(y, y_new)
plt.figure(figsize=(10, 6))
plt.scatter(x_updated, y_updated, label='Updated data points', color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Updated Data Points')
plt.legend()
plt.grid(True)
plt.savefig('updated_data.png')
plt.show()

# Plotting statistics
# Plot x_bar
plt.figure(figsize=(10, 6))
plt.axhline(y=x_bar, label='Initial x_bar', color='blue')
plt.axhline(y=x_bar_new, label='Updated x_bar', color='red')
plt.xlabel('Sample')
plt.ylabel('x_bar')
plt.title('x_bar Over Samples')
plt.legend()
plt.grid(True)
plt.savefig('x_bar.png')
plt.show()

# Plot y_bar
plt.figure(figsize=(10, 6))
plt.axhline(y=y_bar, label='Initial y_bar', color='blue')
plt.axhline(y=y_bar_new, label='Updated y_bar', color='red')
plt.xlabel('Sample')
plt.ylabel('y_bar')
plt.title('y_bar Over Samples')
plt.legend()
plt.grid(True)
plt.savefig('y_bar.png')
plt.show()

# Plot C_xx
plt.figure(figsize=(10, 6))
plt.axhline(y=C_xx, label='Initial C_xx', color='blue')
plt.axhline(y=C_xx_new, label='Updated C_xx', color='red')
plt.xlabel('Sample')
plt.ylabel('C_xx')
plt.title('C_xx Over Samples')
plt.legend()
plt.grid(True)
plt.savefig('C_xx.png')
plt.show()

# Plot C_xy
plt.figure(figsize=(10, 6))
plt.axhline(y=C_xy, label='Initial C_xy', color='blue')
plt.axhline(y=C_xy_new, label='Updated C_xy', color='red')
plt.xlabel('Sample')
plt.ylabel('C_xy')
plt.title('C_xy Over Samples')
plt.legend()
plt.grid(True)
plt.savefig('C_xy.png')
plt.show()
```
